### 1. Setting up Spcay Environment

```python
import spacy
```

```python
# loading the language model for tekenizatoin and stuffs from spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# reading texts from txt file
with open("wiki_fiction.txt", "r", encoding="utf-8") as f:
  text = f.read()
```

```python
print(text)
```

```python
doc = nlp(text)
```

```python
print(doc)
```

### 2. Difference Between normal texts and with using spacy

```python
print("Length of texts read as it is : ",len(text))
print("Length of doc after using spacy : ",len(doc))
```

```python
i=1 
print("token_number:token_character")
for token in text[:20]:
    print(i,end=":")
    print(token,end="  ")
    i+=1
    
```

```python
i=1 
print("token_number:token_character")
for token in doc[:20]:
    print(i,end=":")
    print(token,end="  ")
    i+=1
```

### 3. Sentence Boundary Detection (SBD)

```python
for sentence in list(doc.sents)[:20]:
    print(sentence)
```

```python
sentence1 = list(doc.sents)[1]
print(sentence1)
for token in sentence1:
    print(token)
```

```python
token2 = sentence1[3]
print(token2)
```

```python
# Some token attributes
print(".text :",token2.text)
print(".left_edge :",token2.left_edge)
print(".right_edge :",token2.right_edge)
print(".head :",token2.head)
print(".ent_iob_ :",token2.ent_iob_)

# Printing the base form of the verb using .lemma_
print("\nActual Word :",sentence1[5])
print("using .lemma_ :",sentence1[5].lemma_)
print(".morph :",sentence1[5].morph)

# Printing part of speech
print(".pos_ :",token2.pos_)

# printing dependency relation
print(".dep_ :",token2.dep_)

print(".lang_ :",token2.lang_)
```

### 5. Part of Speech Tagging (POS)

```python
text2 = "Dhruv enjoys playing a PC game that is Valorant."
doc2 = nlp(text=text2)
print(doc2)
```

```python
for token in doc2:
    print(token.text, token.pos_, token.dep_)
```

```python
# visualizing from the library displacy
from spacy import displacy

displacy.render(doc2, style="dep")
```

### 6. Named Entity Recognition

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```
